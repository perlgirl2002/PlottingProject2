```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

This Jupyter Notebook contains multiple examples of data visualizations related to beef production in the United States since 1931.  This information was retrieved from "The National Summary of Meats".  The following has been taken directly from the United States Department of Agriculture for the descriptions of the different types of beef.  

__Quality Grades:__
- **Prime** grade  is produced from young, well-fed beef cattle. It has abundant marbling and is generally sold in restaurants and hotels. Prime roasts and steaks are excellent for dry-heat cooking (broiling, roasting, or grilling).

- **Choice** grade  is high quality, but has less marbling than Prime. Choice roasts and steaks from the loin and rib will be very tender, juicy, and flavorful and are, like Prime, suited to dry-heat cooking. Many of the less tender cuts, such as those from the rump, round, and blade chuck, can also be cooked with dry heat if not overcooked. Such cuts will be most tender if "braised" â€” roasted, or simmered with a small amount of liquid in a tightly covered pan.

- **Select** grade  is very uniform in quality and normally leaner than the higher grades. It is fairly tender, but, because it has less marbling, it may lack some of the juiciness and flavor of the higher grades. Only the tender cuts (loin, rib, sirloin) should be cooked with dry heat. Other cuts should be marinated before cooking or braised to obtain maximum tenderness and flavor.

- **Standard** and **Commercial** grades are frequently sold as ungraded or as "store brand" meat.

- **Utility**, **Cutter**, and **Canner** grades are seldom, if ever, sold at retail but are used instead to make ground beef and processed products.

```python
beef_df = pd.read_csv("C:\\Users\\perlg_000\\data\\beef.csv", na_values=0, thousands=',')
```

The following graph is a scatter plot showing in Millions of Pounds how much beef was graded per year in the United States.  The graph was developed using matplotlib.

```python
plt.scatter(x=beef_df['YEAR'], y=beef_df['POUNDS GRADED'])
plt.ylim(0, 29000)
plt.xlim(1930, 2015)
plt.xlabel("Years")
plt.ylabel("Millions of Pounds")
plt.title("Millions of Pounds of Beef graded per year in the US")
plt.show()
```

The following graph is a scatter plot for the same data above but completed using the pandas built-in plotting functionality.

```python
beef_df.plot(x='YEAR', y='POUNDS GRADED', kind='scatter')
plt.title("Millions of Pounds of Beef graded per year in the US")
```

The following graph is a line plot for the same data above.

```python
plt.plot(beef_df['YEAR'], beef_df['POUNDS GRADED'])
plt.show()
```

The following graph is a line graph comparing the different grades of meat in the Millions of Pounds in the United States since 1931.

```python
beef_df.PRIME = beef_df.PRIME.astype(float)
plt.plot(beef_df['YEAR'], beef_df['PRIME'], label='Prime')
plt.plot(beef_df['YEAR'], beef_df['CHOICE'], label='Choice')
plt.plot(beef_df['YEAR'], beef_df['SELECT'], label='Select')
plt.plot(beef_df['YEAR'], beef_df['COMRCL'], label='Commercial')
plt.plot(beef_df['YEAR'], beef_df['UTILITY'], label='Utility')
plt.plot(beef_df['YEAR'], beef_df['CUTTER'], label='Cutter')
plt.plot(beef_df['YEAR'], beef_df['CANNER'], label='Canner')
plt.legend()
plt.show()
```

```python
beef_df.dropna(subset=['YEAR'], inplace=True)
```

The following graph shows the grade of Choice in Millions of Pounds per Year in the United States but using the built-in plotting functionality from the pandas library.

```python
beef_df.set_index('YEAR')[['CHOICE']].plot()
```

The following graphs show 3 different types of meat grades in matplotlib subplot format.

```python
figure = plt.figure()
ax1 = figure.add_subplot(311)
ax2 = figure.add_subplot(312)
ax3 = figure.add_subplot(313)
figure.subplots_adjust(hspace=.95)

ax1.plot(beef_df['YEAR'], beef_df['CUTTER'], color='green')
ax1.set_title('Millions of Pounds of Cutter Beef graded since 1931')
ax2.plot(beef_df['YEAR'], beef_df['CANNER'], color='indigo', label='Canner')
ax2.set_title('Millions of Pounds of Canner Beef graded since 1931')
ax3.plot(beef_df['YEAR'], beef_df['UTILITY'], color='pink', label='Utility')
ax3.set_title('Millions of Pounds of Utility Beef graded since 1931')
plt.show()
```
